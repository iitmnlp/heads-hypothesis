```python
"""
This computes the block score for the heads.
"""
import pickle as pkl
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as st
import os
import tokenization
import pandas as pd
import itertools
from scipy.stats import entropy

from matplotlib import colors
import matplotlib.pyplot as plt
num_sents = 20
num_layers, num_heads = 12, 12
```

```python
def init(task_name):
    input_file='./texts/glue_data_{}_test_small.tsv'.format(task_name.upper())
    attn_file1='./pkl_dir/attention_weights_{}_small.pkl'.format(task_name)
    task_name=task_name
    
    # Read attention weights file.
    a = pkl.load(open(attn_file1, 'rb'))
    print("len of a {}".format(len(a)))
    print("shape of a {}".format(a[0].shape))
    
    # Read text file.
    f = open(input_file, 'r')
    f1 = f.readlines()
    f1 = f1[1:]   #Ignore the 1st row which has label.
    print("Total sentences in test set: {}".format(len(f1)))
    return f1, a
```

```python
# p-val calculation
def ztest_own(pop1, th, n, alternate='smaller'):
    mu1, v = np.mean(pop1), np.std(pop1) 
    z = (mu1 - th)/(v/np.sqrt(n))
    df = n - 1
    rv = st.t(df)
    if alternate == 'larger':
        return rv.cdf(z)
    else:
        #print('here')
        return 1 - rv.cdf(z)
    
def get_pvalues(uniform_baseline_score, out_band):
    alt_pval = np.zeros([num_layers, num_heads])
    for l in range(num_layers):
            for h in range(num_heads):
                single_head_scores = []
                for s in range(num_sents):
                    single_head_scores.append(out_band[s][l,h])
                if len(uniform_baseline_score) != len(single_head_scores):
                    raise "Population lengths not same for p-value test."
                score = ztest_own(single_head_scores, uniform_baseline_score[0], num_sents)
                alt_pval[l,h] = score
    return alt_pval
```

```python
def calc_block_attn_score(a, f1, tokenizer, task_name):
    if task_name == 'sst-2':
        raise Exception("This task involves only single-sentence.")
    block_score = np.zeros((num_sents, num_layers, num_heads))
    #print(num_sents)
    for i in range(num_sents):
        sent = f1[i].split('\t')
        if task_name in ['qqp', 'qnli']:
            sent_a = tokenizer.tokenize(sent[1])
            sent_b = tokenizer.tokenize(sent[2])
            final_sent = ['CLS '] + sent_a + [' SEP '] + sent_b + [' SEP']
        elif task_name == 'mrpc':
            sent_a = tokenizer.tokenize(sent[3])
            sent_b = tokenizer.tokenize(sent[4])
            final_sent = ['CLS '] + sent_a + [' SEP '] + sent_b + [' SEP']
        elif task_name == 'sst-2':
            sent_a = tokenizer.tokenize(sent[1])
            final_sent = ['CLS '] + sent_a + [' SEP']
        sent_len = len(final_sent)
        if sent_len > max_seq_len:
            sent_len = max_seq_len
            
        cls_idx = 0
        sep1_idx, sep2_idx = len(sent_a)+1, sent_len-1
        
        if task_name in ['sst-2']:
            sep2_idx = -1
    
        for l in range(num_layers):
            for h in range(num_heads):
                attn_per_head = a[i][l,h][:sent_len, :sent_len]
                sent_score = 0
                for tok_id, token in enumerate(range(sent_len)):
                    avg_attn_over_all_toks = np.sum(attn_per_head[token, :])/sent_len
                    if tok_id < sep1_idx:
                        s_size = sep1_idx - cls_idx - 1
                        s_avg_attn_per_tok = np.sum(attn_per_head[token, 1:sep1_idx])/s_size
                    else:
                        s_size = sep2_idx - sep1_idx - 1
                        s_avg_attn_per_tok = np.sum(attn_per_head[token, (sep1_idx+1):sep2_idx])/s_size
                    
                    token_score  = (s_avg_attn_per_tok/avg_attn_over_all_toks)*entropy(attn_per_head[token, :])
#                     if l == 3 and (h == 7 or l == 8):
#                         print("L{}-H{}".format(l,h))
#                         print(entropy(attn_per_head[token, :]))
                    sent_score+=token_score
                sent_score = sent_score/sent_len
                block_score[i,l,h] = sent_score
    return block_score
```

```python
# visualize pval and score map.
def vis_and_save(out_band, task_name, output_dir):
    plt.figure(figsize=(8,6))
    g = sns.heatmap(np.average(out_band, axis=0), cmap='Blues', annot=True, vmin=3)   
    g.set(xticklabels=np.arange(12), yticklabels=np.arange(12))
    plt.xlabel('heads')
    plt.ylabel('layers')
    plt.title('Avg block bias for {}'.format(task_name))
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    pkl.dump(out_band, open(output_dir + "all_heads_all_sents_block_score_{}.pkl".format(task_name), 'wb'))
```

```python
# Driver cell
for task_name in  ['qnli', 'qqp', 'mrpc']: # 'sst-2' is a single-sentence task so does not have block patterns.
    %reset_selective -f "^a$"
    max_seq_len=128
    output_dir = "./sieve_scores/block/"
    tokenizer = tokenization.FullTokenizer(vocab_file="vocab.txt", do_lower_case=True)

    f1, a = init(task_name)
    # No-op heads 
    block_score = calc_block_attn_score(a, f1, tokenizer, task_name)
    vis_and_save(block_score, task_name, output_dir)
```
